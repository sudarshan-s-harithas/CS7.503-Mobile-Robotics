# Question 2:  Pose Graph Optimization for 2D SLAM

## Objective
A robot is travelling in a oval trajectory. It is equipped with wheel odometry for odometry information and RGBD sensors for loop closure information. Due to noise in wheel odometry it generates a noisy estimate of the trajectory. Our task is to use loop closure pairs to correct the drift.

We pose this problem as a graph optimization problem. In our graph, poses are the vertices and constraints are the edges. 

## Given: 
In practical scenarios, we'd obtain the following from our sensors after some post-processing:

1. Initial position
2. Odometry Contraints/Edges: This "edge" information tells us relative transformation between two nodes. These two nodes are consecutive in the case of Odometry but not in the case of Loop Closure (next point).
3. Loop Closure Contraints/Edges: Remember that while optimizing, you have another kind of "anchor" edge as you've seen in 1. solved example.

You have been given a text file named `edges.txt` (in `data/`) which has all the above 3 and it follows G2O's format (as explained in class, [link here](https://saishubodh.notion.site/G2O-Edge-Types-d9f9ff63c77c4ceeb84b1e49085004e3) ). The ground truth is `gt.txt`.

## Steps:

### Part 1

Using the following motion model, you have to first generate the "initialization" for all the poses/vertices using the "Given" information. Just like in the 1D case.
$$x_{k+1} = x_{k} + \Delta x_{(k,k+1)} \cos(\theta_k) - \Delta y_{(k,k+1)} \sin(\theta_k) \\
y_{k+1} = y_{k} + \Delta y_{(k,k+1)} \cos(\theta_k) + \Delta x_{(k,k+1)} \sin(\theta_k) \\
\theta_{k+1} = \theta_{k}+  \Delta \theta_{(k,k+1)} \tag{3}$$

Even the loop closure nodes are related by the above model, except that it need not necessarily be consecutive notes k and k+1.

Save this initial trajectory as `edges-poses.g2o`.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from jax import jacfwd, ops
import jax.numpy as jnp 


def readEdge(fileName):
    f = open(fileName, 'r')
    A = f.readlines()
    f.close()
    index_arr_i = []
    index_arr_j = []
    edge_x = []
    edge_y = []
    edge_theta = []
    for line in A:
        if "EDGE_SE2" in line:
            (edge, ind1, ind2, dx, dy, dtheta, _, _, _, _, _, _) = line.split()
            index_arr_i.append(int(ind1))
            index_arr_j.append(int(ind2))
            edge_x.append(float(dx))
            edge_y.append(float(dy))
            edge_theta.append(float(dtheta))
            
    return (np.array( index_arr_i), np.array(index_arr_j), np.array(edge_x), np.array(edge_y), 
            np.array(edge_theta))

def readVertex(fileName):
    f = open(fileName, 'r')
    A = f.readlines()
    f.close()
    x_arr = []
    y_arr = []
    theta_arr = []
    for line in A:
        if "VERTEX_SE2" in line:
            (ver, ind, x, y, theta) = line.split()
            x_arr.append(float(x))
            y_arr.append(float(y))
            theta_arr.append(float(theta.rstrip('\n')))

    return np.array([x_arr, y_arr, theta_arr])

def drawThree(X1, Y1, THETA1, X2, Y2, THETA2, X3, Y3, THETA3):
    
    ax = plt.subplot(111)
    ax.plot(X1, Y1, 'ro', label='Ground Truth')
    plt.plot(X1, Y1, 'k-')
    for i in range(len(THETA1)):
        x2 = 0.25*math.cos(THETA1[i]) + X1[i]
        y2 = 0.25*math.sin(THETA1[i]) + Y1[i]
        plt.plot([X1[i], x2], [Y1[i], y2], 'r->')
    ax.plot(X2, Y2, 'bo', label='Optimized')
    plt.plot(X2, Y2, 'k-')
    for i in range(len(THETA2)):
        x2 = 0.25*math.cos(THETA2[i]) + X2[i]
        y2 = 0.25*math.sin(THETA2[i]) + Y2[i]
        plt.plot([X2[i], x2], [Y2[i], y2], 'b->')
    ax.plot(X3, Y3, 'go', label='Noisy')
    plt.plot(X3, Y3, 'k-')
    for i in range(len(THETA3)):
        x2 = 0.25*math.cos(THETA3[i]) + X3[i]
        y2 = 0.25*math.sin(THETA3[i]) + Y3[i]
        plt.plot([X3[i], x2], [Y3[i], y2], 'g->')
    plt.legend()
    plt.show()
    
def draw(X, Y, THETA):
    ax = plt.subplot(111)
    ax.plot(X, Y, 'ro')
    plt.plot(X, Y, 'c-')
    for i in range(len(THETA)):
        x2 = 0.25*np.cos(THETA[i]) + X[i]
        y2 = 0.25*np.sin(THETA[i]) + Y[i]
        plt.plot([X[i], x2], [Y[i], y2], 'g->')
    plt.show()
```

```python
def Update_state_incremental(Jacobian ,omega , error , lambda_val  ):
    H = Jacobian.T@ omega @Jacobian  + np.eye(120)*lambda_val
    b = Jacobian.T@ omega@(-error )
    delta = np.linalg.inv(H)@b
    state_error = (1/2)*(error.T)@omega@(error)
    
    return delta , state_error

def determine_information_matrix(I, J, loop, normal, anchor):
    omega = np.zeros(((len(I)), (len(I))))
    print( np.shape(omega))
    
    for i in range( (len(I) -1 ) ):
        if( J[i] == I[i] +1 ):
            omega[i][i] = normal
        else:
            omega[i][i] = loop
    omega[len(I) -1][len(I)-1] = anchor
    
    return omega

def get_error(i, j, X_pose, Y_pose, theta_pose ,X_edge,Y_edge,theta_edge):
    error_temp_x = np.zeros( (140, 1) )
    error_temp_y = np.zeros( (140, 1) )
    error_temp_theta = np.zeros( (140, 1) )
    
    for p in range( len(I) ):
        errorx =  X_pose[i[p]] + (X_edge[i[p]])*np.cos(theta_pose[i[p]]) - (Y_edge[i[p]])*np.sin(theta_pose[i[p]]) -X_pose[j[p]]
        errory = Y_pose[i[p]] + (Y_edge[i[p]] )*np.cos(theta_pose[i[p]] ) + (X_edge[i[p]])*np.sin(theta_pose[i[p]]) - Y_pose[j[p]]
        error_theta = theta_pose[i[p]] + (theta_edge[i[p]]) - theta_pose[j[p]]
        
        error_temp_x[p][0] = errorx
        error_temp_y[p][0] = errory
        error_temp_theta[p][0] = error_theta

    return error_temp_x , error_temp_y , error_temp_theta
```

```python
file_path = "../edges.txt"
file_path_gt = "../gt.txt"
X_init,Y_init,theta_init = readVertex(file_path)
I,J,X_edge,Y_edge,theta_edge = readEdge(file_path)
X_gt , Y_gt, Theta_gt = readVertex(file_path_gt)

X_edge = np.append(X_edge,0)
Y_edge = np.append(Y_edge,0)
theta_edge = np.append(theta_edge,0)
I = np.append(I,0)
J = np.append(J,0)
X_edge = X_edge.reshape(140,1)
Y_edge= Y_edge.reshape(140,1)
theta_edge = theta_edge.reshape(140,1)
```

```python
Jacobian = np.zeros(((len(I)), len(X_gt))) ## [140,120] [ (number_of_edges +anchor) X (number_of_states_or_vertex) ]
X_poses = np.zeros(((len(I)), 1))
Y_poses = np.zeros(((len(I)), 1))
Theta_poses = np.zeros(((len(I)), 1))
cnt =1

X_poses[0][0] = X_init[0]
Y_poses[0][0] = Y_init[0]
Theta_poses[0][0] = theta_init[0]
c = np.cos
s = np.sin

Jacobian[0][0] = 1
Jacobian[0][1] = -1

for i in range( len(I)-1  ): ## i will go from  0 till 139
    X_poses[cnt] = X_poses[ I[i] ][0] + X_edge[i]*c( Theta_poses[i] ) - Y_edge[i]*s(Theta_poses[i])
    Y_poses[cnt] = Y_poses[ I[i] ][0]  + Y_edge[i]*c( Theta_poses[i] )  +X_edge[i]*s( Theta_poses[i] )
    Theta_poses[cnt] = Theta_poses[ I[i] ][0] + theta_edge[i]
    pre_index = I[i]  
    post_index = J[i]
    cnt +=1
    Jacobian[i][post_index] = -1
    Jacobian[i][pre_index] = 1
    
Jacobian[139][0] =1 ## anchor to origin 
draw( X_poses[0:119] ,Y_poses[0:119] , Theta_poses[0:119])
```

```python
X_state = X_poses[0:119]
Y_state = Y_poses[0:119]
Theta_state = Theta_poses[0:119]

X_state = np.append( X_state ,  X_init)
Y_state = np.append( Y_state ,  Y_init)
Theta_state = np.append( Theta_state ,  theta_init)

X_state = np.reshape(X_state , ( 120, 1))
Y_state = np.reshape(Y_state , ( 120, 1))
Theta_state = np.reshape(Theta_state , ( 120, 1))

error_x , error_y , error_theta = get_error(I, J, X_state,
                                            Y_state, Theta_state,
                                            X_edge,Y_edge,theta_edge)

loop ,normal ,  anchor = 10 , 2 , 20
omega = determine_information_matrix(I, J , loop ,normal ,  anchor )
```

```python
error_x, error_y, error_theta = get_error(I, J, X_state,
                                          Y_state, Theta_state,
                                          X_edge,Y_edge,theta_edge)
lambdx = 10**-2
lambdy = 10**-2
lambdt = 10**-2

Tj = Theta_state
Yj = Y_state
Xj = X_state

X_state_update , Y_state_update , Theta_state_update = X_state , Y_state , Theta_state
temp_state_X ,  temp_state_Y , temp_state_theta = X_state , Y_state , Theta_state

tempx = X_state
tempy = Y_state
tempt = Theta_state
xx , yy,tt  =X_gt , Y_gt, Theta_gt 

for m in range(50):
    
    delta_x , errorx = Update_state_incremental( Jacobian , omega ,  error_x , lambdx )
    delta_y , errory =  Update_state_incremental( Jacobian , omega ,  error_y , lambdy )
    delta_t , errort  =  Update_state_incremental( Jacobian , omega ,  error_theta , lambdt )

    temp_state_X = temp_state_X + delta_x
    temp_state_Y = temp_state_Y + delta_y
    temp_state_theta = temp_state_theta + delta_t

    fx0_temp, fy0_temp, ft0_temp = get_error(I, J, temp_state_X,
                                             temp_state_Y, temp_state_theta,
                                             X_edge,Y_edge,theta_edge)
    
    error_tempx  = (0.5) * (fx0_temp.T @ omega @ fx0_temp)
    error_tempy  = (0.5) * (fy0_temp.T @ omega @ fy0_temp)
    error_tempt  = (0.5) * (ft0_temp.T @ omega @ ft0_temp)

    if (error_tempt < errort):
        Theta_state_update =  Theta_state_update + delta_t
        lambdt = lambdt / 10
    else:
        lambdt = lambdt * 10

    if (error_tempy <  errory):
        Y_state_update =  Y_state_update + delta_y
        lambdy = lambdy / 10
    else:
        lambdy = lambdy * 10
        
    if (error_tempx < errorx):
        X_state_update =  X_state_update +  delta_x
        lambdx = lambdx / 10
    else:
        lambdx = lambdx * 10

    error_x, error_y, error_theta = get_error(I, J, X_state_update,
                                              Y_state_update , Theta_state_update,
                                              X_edge,Y_edge,theta_edge)    
    error = (errorx + errory + errort)/3
    
    if m%10 == 0:
        print( "Iteration Number = " + str(m ) + " " "Error = "+  str(error[0][0])  ) 
        drawThree( X_gt , Y_gt, Theta_gt, 
                  X_state_update, Y_state_update ,  Theta_state_update , 
                  X_poses[0:119] ,Y_poses[0:119] , Theta_poses[0:119])
```

```python
def writeOdom(X, Y, THETA):
    g2o = open('optimized_pose.g2o', 'w')
    for i, (x, y, theta) in enumerate(zip(X,Y,THETA)):
        line = "VERTEX_SE2 " + str(i) + " " + str(x[0]) + " " + str(y[0]) + " " + str(theta[0])
        g2o.write(line)
        g2o.write("\n")

    info_mat = "500.0 0.0 0.0 500.0 0.0 500.0"
    for i in range(1, len(X)):
        p1 = (X[i-1][0] , Y[i-1][0] , THETA[i-1][0] )
        p2 = (X[i][0] , Y[i][0] , THETA[i][0] )
        T1_w = np.array([[np.cos(p1[2]), -np.sin(p1[2]), p1[0]], [np.sin(p1[2]), np.cos(p1[2]), p1[1]], [0, 0, 1]])
        T2_w = np.array([[np.cos(p2[2]), -np.sin(p2[2]), p2[0]], [np.sin(p2[2]), np.cos(p2[2]), p2[1]], [0, 0, 1]])
#         T1_w += np.eye( np.shape(T1_w)[0] )*0.01
        T2_1 = np.dot(np.linalg.inv(T1_w), T2_w)
#         T2_1 = np.dot(T1_w.T, T2_w)
        
        del_x = str(T2_1[0][2])
        del_y = str(T2_1[1][2])
        del_theta = str(math.atan2(T2_1[1, 0], T2_1[0, 0]))

        line = "EDGE_SE2 "+str(i-1)+" "+str(i)+" "+del_x+" "+del_y+" "+del_theta+" "+info_mat
        g2o.write(line)
        g2o.write("\n")

    g2o.write("FIX 0")
    g2o.write("\n")
#     g2o.close()
    return g2o
```

### Comparision of Jacobain using Jax

```python
def diff_x(X_pose,X_edge,Y_edge,theta_pose,i,j):
    res_x =  X_pose[i] + (X_edge[i])*jnp.cos(theta_pose[i]) - (Y_edge[i])*jnp.sin(theta_pose[i]) - X_pose[j]
    
    return res_x
```

```python
def diff_y(Y_pose,X_edge,Y_edge,theta_pose,i,j):
    res_y =  Y_pose[i] + (Y_edge[i])*jnp.cos(theta_pose[i]) + (X_edge[i])*jnp.sin(theta_pose[i]) - Y_pose[j] 
    
    return res_y
```

```python
def diff_theta(theta_pose,theta_edge,i,j  ):
    
    res_theta = theta_pose[i] + (theta_edge[i]) - theta_pose[j] 
    
    return res_theta
```

```python
def evaluate_jacobain(  J_analytical, X_pose, Y_pose, theta_pose,
                      X_edge , Y_edge , Theta_edge ,
                      I, J ):
    
    func_x = lambda x : diff_x(x,jnp.array(X_edge),jnp.array(Y_edge),jnp.array(theta_pose),jnp.array(I),jnp.array(J))
    func_y = lambda y : diff_y(y,jnp.array(X_edge),jnp.array(Y_edge),jnp.array(theta_pose),jnp.array(I),jnp.array(J))
    func_theta = lambda t : diff_theta(t,jnp.array(theta_edge),I,J)
    
    
    Jx = jacfwd(func_x)
    Jy = jacfwd(func_y)
    Jt = jacfwd(func_theta)
    
    JJx = Jx(jnp.array(X_state_update))
    JJy = Jy(jnp.array(Y_state_update))
    JJt = Jt(jnp.array(Theta_state_update))
    
    JJx =JJx.reshape(140,120)
    JJy =JJy.reshape(140,120)
    JJt =JJt.reshape(140,120)
     
    JJx = ops.index_update(JJx, ops.index[139,0], 1)
    JJy = ops.index_update(JJy, ops.index[139,0], 1)
    JJt = ops.index_update(JJt, ops.index[139,0], 1)
    
    jnp.set_printoptions(suppress=True)
    
    val_x = jnp.linalg.norm(JJx - J_analytical, 'fro') ## Calculate the norm 
    val_y = jnp.linalg.norm(JJy - J_analytical, 'fro')
    val_theta = jnp.linalg.norm(JJt - J_analytical, 'fro')
    
    print( " The result of the norm of difference between jax and analytical Jacobian  for X is  " + str(val_x)   )
    print( " The result of the norm of difference between jax and analytical Jacobian  for Y is  " + str(val_y)   )
    print( " The result of the norm of difference between jax and analytical Jacobian  for Theta is  " + str(val_theta)   )
    
    
```
