## Worked out Example using Gradient Descent

A Gaussian distribution parametrized by $a,m,s$ is given by:

$$ y(x;a,m,s)=a \exp \left(\frac{-(x-m)^{2}}{2 s^{2}}\right) \tag{1}$$

### Jacobian of Gaussian

$$\mathbf{J}_y=\left[\frac{\partial y}{\partial a} \quad \frac{\partial y}{\partial m} \quad \frac{\partial y}{\partial s}\right] \\
= \left[ \exp \left(\frac{-(x-m)^{2}}{2 s^{2}}\right); \frac{a (x-m)}{s^2} \exp\left(\frac{-(x-m)^{2}}{2 s^{2}}\right);  \frac{a (x-m)^2}{s^3}\exp \left(\frac{-(x-m)^{2}}{2 s^{2}}\right)\right]$$

## Problem at hand

> Given a set of observations $y_{obs}$ and $x_{obs}$ we want to find the optimum parameters $a,m,s$ which best fit our observations given an initial estimate.

Our observations would generally be erroneous and given to us, but for the sake of knowing how good our model is performing, let us generate the observations ourselves by assuming the actual "actual" parameter values as $a_{gt}=10; m_{gt} =0; s_{gt} =20$ ($gt$ stands for ground truth). We will try to estimate these values based on our observations and let us see how close we get to "actual" parameters. Note that in reality we obviously don't have these parameters as that is exactly what we want to estimate in the first place. So let us consider the following setup, we have:

- Number of observations, $num\_obs = 50$
- Our 50 set of observations would be
    - $x_{obs} = np.linspace(-25,25, num\_obs)$
    - $y_{obs} = y(x_{obs};a_{gt},m_{gt},s_{gt})$  from $(1)$

Reference:

â†’[linspace](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html)

- Say we are given initial estimate as:

    $$a_0=10; \quad m_0=13; \quad s_0=19.12$$

### Residual and error to be minimized

Okay, now we have set of observations and an initial estimate of parameters. We would now want to minimize an error that would give us optimum parameters.

The $residual$ would be given by

$$ r(a,m,s) = \left[ a \exp \left(\frac{-(x_{obs}-m)^{2}}{2 s^{2}}\right) - y_{obs}\ \right]$$

where we'd want to minimize $\|r\|^2$. Note that $r$ is a non-linear function in $(a,m,s)$.

Also, note that since $y$ (and $x$) are observations in the above equation, after simplification, we get $\mathbf{J}_r = \mathbf{J}_y$ [above](https://www.notion.so/c9e6f71b67a44bb8b366df2fccfc12d0) (since $y_{obs}$ is a constant).

Let us apply Gradient Descent method for minimization here. From [Table I](https://www.notion.so/From-linear-algebra-to-non-linear-weighted-least-squares-optimization-13cf17d318be4d45bb8577c4d3ea4a02),  

$$\Delta \mathbf{k} = - \alpha \mathbf{J_F} = -\alpha \mathbf{J}_r^{\top} {r}(\mathbf{k})$$

Note that $\mathbf{J_F}$ is the Jacobian of "non-linear least squares" function $\mathbf{F}$ while $\mathbf{J}_r$ is the Jacobian of the residual. 

where $\mathbf{k}$ is $[a,m,s]^T$. 

- Some hyperparameters:
    - Learning rate, $lr = 0.01$
    - Maximum number of iterations, $num\_iter=200$
    - Tolerance, $tol = 1e-15$

## Solution for 1 iteration

To see how each step looks like, let us solve for 1 iteration and for simpler calculations, assume we have 3 observations, 

$$x_{obs}= \left[ -25, 0, 25 \right]^T, y_{obs} = \left[  4.5783, 10, 4.5783 \right]^T. $$

With our initial estimate as $\mathbf{k_0} = [a_0=10, \quad m_0=13, \quad s_0=19.12]^T$, the residual would be 

$$ r(a_0,m_0,s_0) = \left[ a_0 \exp \left(\frac{-(x_{obs}-m_0)^{2}}{2 s_0^{2}}\right) - y_{obs}\ \right]$$

Therefore, $r=[-3.19068466, -2.0637411 , 3.63398058]^T$.

### Gradient Computation

Gradient, $\mathbf{J_F}$=

$$\mathbf{J_r}^{\top} \mathbf{r}(\mathbf{k})$$

We have calculated residual already [above](https://www.notion.so/c9e6f71b67a44bb8b366df2fccfc12d0), let us calculate the Jacobian $\mathbf{J_r}$.

$$\mathbf{J}_r
= \left[ \exp \left(\frac{-(x-m)^{2}}{2 s^{2}}\right); \frac{a (x-m)}{s^2} \exp\left(\frac{-(x-m)^{2}}{2 s^{2}}\right);  \frac{a (x-m)^2}{s^3}\exp \left(\frac{-(x-m)^{2}}{2 s^{2}}\right)\right]$$

$$\implies \mathbf{J_r} = \left[ \begin{array}{rrr}0.1387649 & 0.79362589, & 0.82123142 \\-0.14424057 & -0.28221715  & 0.26956967 \\0.28667059 & 0.19188405, & 0.16918599\end{array}\right]$$

So ,

$$\mathbf{J_F} = \mathbf{J_r}^{\top} \mathbf{r}(\mathbf{k})$$

$$\mathbf{r}(\mathbf{k}) =  \left[ \begin{array}{r}-3.19068466 \\ -2.0637411 \\ 3.63398058 \end{array} \right]$$

$$ \begin{aligned} \implies \mathbf{J_F} = \left[ \begin{array}{r} 0.89667553 \\ -1.25248392 \\-2.56179392\end{array} \right] \end{aligned}$$

### Update step

$$
\Delta \mathbf{k} = - \alpha \mathbf{J_F} \\
\mathbf{k}^{t+1} = \mathbf{k}^t + \Delta \mathbf{k}
$$

Here, $\alpha$ our learning rate is 0.01.

$$
\Delta \mathbf{k} = - \alpha\times\left[ \begin{array}{r} 
0.89667553 \\ -1.25248392 \\-2.56179392
\end{array} \right] = \left[ \begin{array}{r}
-0.00896676 \\ 0.01252484 \\0.02561794
\end{array}\right]
$$

$$
\mathbf{k}^{1} = \mathbf{k}^{0} + \Delta \mathbf{k} \\ \left[\begin{array}{r} 10 \\ 13 \\ 19.12 \end{array}\right] + \left[\begin{array}{c} 9.99103324 \\ 13.01252484 \\ 19.14561794 \end{array} \right]
$$

With just one iteration with very few observations, we can see that we have gotten *slightly* more closer to our GT parameter  $a_{gt}=10; m_{gt} =0; s_{gt} =20$. ## 1.3: Different Optimizers

Replace gradient descent with Gauss-Newton and Levenberg Marquardt algorithms and repeat question 1.1. 

To quickly recap, Gauss-Newton and Levenberg Marquardt are alternate update rules to the standard gradient descent. Gauss Newton updates work as:

$$\delta x = -(J^TJ)^{-1}J^Tf(x)$$

Levenberg Marquardt lies somewhere between Gauss Newton and Gradient Descent algorithms by blending the two formulations. As a result, when at a steep cliff, LM takes small steps to avoid overshooting, and when at a gentle slope, LM takes bigger steps:


$$\delta x = -(J^TJ + \lambda I)^{-1}J^Tf(x)$$

#### Solution 1.1, 1.3 : Non Linear Least Squares Optimization for first function using Gradient Decent, Guass Newton and LM method  

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def generate_samples(x_obs , gt_parameters):
    samples = np.zeros( ( len(x_obs) , 1  ) )
    cnt =0
    for obs in ( x_obs): 
        samples[cnt] = gt_parameters[0]* ( np.exp( -( obs- gt_parameters[1] )**2  / (2* gt_parameters[2]**2))  )
        cnt +=1      
    return samples  

def determine_jacobian(samples , parameters):
    J = np.zeros( ( len(samples) , 3)  )
    A = np.zeros( ( len(samples) , 1)  )
    B = np.zeros( ( len(samples) , 1)  )
    C = np.zeros( ( len(samples) , 1)  )
    cnt = 0
    for obs in samples: 
        A[cnt][0] = ( np.exp( -( obs- parameters[1] )**2  / (2* parameters[2]**2))  )
        B[cnt][0] = (parameters[0] *( obs - parameters[1] )/ parameters[2]**2 ) * (  np.exp( -( obs- parameters[1] )**2  / (2* parameters[2]**2)) )
        C[cnt][0] = (parameters[0] *( obs - parameters[1] )**2/ parameters[2]**3 ) * (  np.exp( -( obs- parameters[1] )**2  / (2* parameters[2]**2)) )
        J[cnt][0] = A[cnt][0]
        J[cnt][1] = B[cnt][0]
        J[cnt][2] = C[cnt][0]
        cnt +=1
    return J


print("Gradient Decent Algorithm")
num_obs = 50
mu_gt = 0
sigma_gt = 20
a_gt  =  10
gt_parameters = np.array([ a_gt , mu_gt , sigma_gt ] ) # a, m, s
x_obs = np.linspace( -25 , 25 , num_obs )
y_obs = generate_samples(x_obs , gt_parameters)
parameters = np.zeros( (3,1 ))
parameters[0][0] =  10 
parameters[1][0] =   13 
parameters[2][0] =  19.2 
lr = 0.0125
iterations = 950

residual_vector = [] 
for p in range(iterations): 

    curr_samples = generate_samples(x_obs ,  parameters )
    error_vector = curr_samples - y_obs
#     error_vector = np.square(error_vector)
#     error_vector = np.abs(error_vector)
    residual = np.linalg.norm( error_vector)
    J = determine_jacobian(curr_samples ,parameters  )
    del_parameters = (J.T)@error_vector
    del_parameters = -lr*del_parameters
    parameters +=del_parameters 
    residual_vector.append( residual)
           
plt.plot( np.arange( 0 ,iterations  ) , residual_vector )
plt.xlabel("loss")
plt.ylabel("iterations")
plt.title("loss vs number of iterations for Gradient Decent")
plt.show()    
plt.plot(x_obs, y_obs , label = 'Ground Truth')
plt.plot(x_obs, curr_samples , label = 'Predicted')
plt.title("predicted vs GT gaussian by Gradient Decent")
plt.legend()
plt.show()   
print("parameters")
print( parameters)


print("")
print("----------------------------------------------------------------")
print("")

print("Gauss Newton Method")
## Gauss Newton Method
num_obs = 50
mu_gt = 0
sigma_gt = 20
a_gt  =  10
gt_parameters = np.array([ a_gt , mu_gt , sigma_gt ] ) # a, m, s
x_obs = np.linspace( -25 , 25 , num_obs )
y_obs = generate_samples(x_obs , gt_parameters)
parameters = np.zeros( (3,1 ))
parameters[0][0] =  10 
parameters[1][0] =   13 
parameters[2][0] =  19.2 
lr = 0.0115
iterations = 100
residual_vector = [] 

for p in range(iterations):   
    curr_samples = generate_samples(x_obs ,  parameters )
    error_vector = curr_samples - y_obs
    residual = np.linalg.norm( error_vector)
    J = determine_jacobian(curr_samples ,parameters  )
    del_parameters = np.linalg.inv( (J.T)@J)@(J.T)@error_vector
    del_parameters = -lr*del_parameters
    parameters +=del_parameters 
    residual_vector.append( residual)
    
plt.plot(np.arange( 0 ,iterations , 1  ) , residual_vector )
plt.xlabel("loss")
plt.ylabel("iterations")
plt.title("loss vs number of iterations for Gauss Newton")
plt.show()    
plt.plot(x_obs, y_obs , label = 'Ground Truth')
plt.plot(x_obs, curr_samples , label = 'Predicted')
plt.title("predicted vs GT gaussian by Gauss Newton")
plt.legend()
plt.show()   
print("parameters")
print( parameters)

print("")
print("----------------------------------------------------------------")
print("")
print("LM Algorithm")

# LM Algorithm 
num_obs = 50
mu_gt = 0
sigma_gt = 20
a_gt  =  10
gt_parameters = np.array([ a_gt , mu_gt , sigma_gt ] ) # a, m, s
x_obs = np.linspace( -25 , 25 , num_obs )
y_obs = generate_samples(x_obs , gt_parameters)
parameters = np.zeros( (3,1 ))
parameters[0][0] =  10 
parameters[1][0] =   13 
parameters[2][0] =  19.2 
lr = 0.016
iterations = 100
residual_vector = [] 
lambda_val = 10**-3
prev_residual = 10000000000

for p in range(iterations):
    
    curr_samples = generate_samples(x_obs ,  parameters )
    error_vector = curr_samples - y_obs
    residual = np.linalg.norm( error_vector)
    if(residual > prev_residual):
        lambda_val *= 10 
        continue 
    if( residual < prev_residual ):
        lambda_val /= 10    
        prev_residual = residual
    J = determine_jacobian(curr_samples ,parameters  )
    del_parameters = np.linalg.inv( (J.T)@J +  lambda_val*np.eye(3)   )@(J.T)@error_vector
    del_parameters = -lr*del_parameters
    parameters +=del_parameters 
    residual_vector.append( residual)
    
plt.plot( np.arange( 0 ,len(residual_vector) , 1  ) , residual_vector )
plt.xlabel("loss")
plt.ylabel("iterations")
plt.title("loss vs number of iterations for LM")
plt.show()   
plt.plot(x_obs, y_obs , label = 'Ground Truth')
plt.plot(x_obs, curr_samples , label = 'Predicted')
plt.title("predicted vs GT gaussian by LM")
plt.legend()
plt.show() 
print("parameters")
print( parameters)
```

#### Solution 1.2 : Non Linear Least Squares Optimization for first function using Gradient Decent, Guass Newton and Lm method  

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def generate_samples(x_obs , gt_param):
    samples = np.zeros((len(x_obs), 1))
    cnt = 0
    for obs in (x_obs):
        samples[cnt][0] = gt_param[0]*(np.exp(-obs/gt_param[1])) + gt_param[2]*(np.sin(obs/gt_param[3])) 
        cnt += 1  
    return samples

def determine_jacobian(samples , params): 
    J = np.zeros((len(samples) , 4))
    cnt = 0
    for obs in samples:  
        J[cnt][0] = np.exp(-obs/params[1])
        J[cnt][1] = params[0]*np.exp(-obs/params[1])*(obs/params[1]**2)
        J[cnt][2] = np.sin(obs/params[3])
        J[cnt][3] = params[2]*np.cos(-obs/params[3])*(-obs/params[3]**2)                                                           
        cnt +=1 
    return J

## Gradient Descent Code 
num_obs = 50
p1 = 2
p2 = 8
p3 = 4
p4 = 8
gt_parameters = np.array([ p1, p2, p3, p4]) 
x_obs = np.linspace( -20, 30, num_obs)
y_obs = generate_samples(x_obs , gt_parameters) 
parameters = np.zeros( (4, 1))
parameters[0][0] =  1.8
parameters[1][0] =  8.2 
parameters[2][0] =  4.2 
parameters[3][0] =  7.8 
lr = 0.001
iterations = 50
residual_vector = [] 
                                                                             
for p in range(iterations):
#     lr = lr/(p + 1)
    curr_samples = generate_samples(x_obs,  parameters)
    error_vector = curr_samples - y_obs
    residual = np.linalg.norm(error_vector)
    J = determine_jacobian(curr_samples ,parameters)
    del_parameters = (J.T)@error_vector                        
    del_parameters = -lr * del_parameters
    parameters += del_parameters 
    residual_vector.append( residual)
        
plt.plot( np.arange(0, iterations) , residual_vector )
plt.xlabel("loss")
plt.ylabel("iterations")
plt.title("loss vs number of iterations for Gradient Decent")
plt.show()    
plt.plot(x_obs, y_obs , label = 'Ground Truth')
plt.plot(x_obs, curr_samples , label = 'Predicted')
plt.title("predicted vs GT gaussian by Gradient Decent")
plt.legend()
plt.show()   
print("parameters")
print( parameters)

print("Gauss Newton Method")
## Gauss Newton Method
num_obs = 50
p1 = 2
p2 = 8
p3 = 4
p4 = 8
gt_parameters = np.array([ p1, p2, p3, p4]) 
x_obs = np.linspace( -20, 30, num_obs)
y_obs = generate_samples(x_obs , gt_parameters) 
parameters = np.zeros( (4, 1))
parameters[0][0] =  1.8
parameters[1][0] =  8.2 
parameters[2][0] =  4.2 
parameters[3][0] =  7.8 
lr = 0.001
iterations = 100
residual_vector = []  

for p in range(iterations):   
    curr_samples = generate_samples(x_obs ,  parameters )
    error_vector = curr_samples - y_obs
    residual = np.linalg.norm( error_vector)
    J = determine_jacobian(curr_samples ,parameters  )
    del_parameters = np.linalg.inv( (J.T)@J)@(J.T)@error_vector
    del_parameters = -lr*del_parameters
    parameters +=del_parameters 
    residual_vector.append( residual)
    
plt.plot(np.arange( 0 ,iterations , 1  ) , residual_vector )
plt.xlabel("loss")
plt.ylabel("iterations")
plt.title("loss vs number of iterations for Gauss Newton")
plt.show()    
plt.plot(x_obs, y_obs , label = 'Ground Truth')
plt.plot(x_obs, curr_samples , label = 'Predicted')
plt.title("predicted vs GT gaussian by Gauss Newton")
plt.legend()
plt.show()   
print("parameters")
print( parameters)

print("")
print("----------------------------------------------------------------")
print("")
print("LM Algorithm")

# LM Algorithm 
num_obs = 50
p1 = 2
p2 = 8
p3 = 4
p4 = 8
gt_parameters = np.array([ p1, p2, p3, p4]) 
x_obs = np.linspace( -20, 30, num_obs)
y_obs = generate_samples(x_obs , gt_parameters) 
parameters = np.zeros( (4, 1))
parameters[0][0] =  1.8
parameters[1][0] =  8.2 
parameters[2][0] =  4.2 
parameters[3][0] =  7.8 
lr = 0.001
iterations = 100
residual_vector = [] 
lambda_val = 10**-3
prev_residual = 10000000000

for p in range(iterations):
    
    curr_samples = generate_samples(x_obs ,  parameters )
    error_vector = curr_samples - y_obs
    residual = np.linalg.norm( error_vector)
    if(residual > prev_residual):
        lambda_val *= 10 
        continue 
    if( residual < prev_residual ):
        lambda_val /= 10    
        prev_residual = residual
    J = determine_jacobian(curr_samples ,parameters  )
    del_parameters = np.linalg.inv( (J.T)@J +  lambda_val*np.eye(4)   )@(J.T)@error_vector
    del_parameters = -lr*del_parameters
    parameters +=del_parameters 
    residual_vector.append( residual)
    
plt.plot( np.arange( 0 ,len(residual_vector) , 1  ) , residual_vector )
plt.xlabel("loss")
plt.ylabel("iterations")
plt.title("loss vs number of iterations for LM")
plt.show()   
plt.plot(x_obs, y_obs , label = 'Ground Truth')
plt.plot(x_obs, curr_samples , label = 'Predicted')
plt.title("predicted vs GT gaussian by LM")
plt.legend()
plt.show() 
print("parameters")
print( parameters)
```

### 1.3  (2). Do you notice any difference between the three optimizers ?

#### Gradient Decent, Gauss Newton, LM method comparison

The Gradient descent method requires the highest number of iteratiations followed by gauss newton and LM to converge to the solution. \
The gauss newton method utilizes the curvature information in addition to the slope to determine the next step, if the curvature is high, the step is small and for a small curvature the step is high. for this reason it converges faster in comparision to gradient descent. \
The Levenberg-Marquardt curve-fitting method is actually a combination of the two other minimization methods: the gradient descent method and the Gauss-Newton method. In the LM algorithm the error will never increase. The Levenberg-Marquardt method acts more like a gradient-descent method when the parameters are far from their optimal value, and acts more like the Gauss-Newton method when the parameters are close to their optimal value.

## 2.2: ICP alignment

1. Write a function that takes two point clouds as input without known correspondences and perform the iterative closest point algorithm.
2. Perform the ICP alignment between the two bunnies and plot their individual coordinate frames as done in class.
3. Does ICP always give the correct alignment? Why or Why not?
4. What are other variants of ICP and why are they helpful (you can look at point to plane ICP)?

##### Solution 3. 
ICP may not give correct alignment due to following reasons:
1. Insufficient or Incorrect Data Association
2. Inconsitancy in point reading, reading exactly same point every time not possibble. 
3. Sensor noise in depth reading.
4. Corresponding frames are too far.

##### Solution 4. 
We may classify these variants and sub-variants:
1. Selection of point set \
    1.1 Use all points \
    1.2 Uniform Sub sampling \
    1.3 Random sampling \
    1.4 Feature based sampling \
    1.5 Normal-space sampling
2. Matching of point set (Data Association) \
    2.1 Closest point \
    2.2 Normal shooting \
    2.3 Point-to-plane \
    2.4 Projection based approach 
3. Weightage for the corresponding points \
    3.1 Constant weight \
    3.2 Assigning lower weights to pairs with greater point-to-point distances \
    3.3 Weighting based on compatibility of normals \
    3.4 Weighting based on the expected effect of scanner noise on the uncertainty in the error metric
4. Rejecting outlier points \
    4.1 Rejection of corresponding points more than a specified distance apart. \
    4.2 Rejection of the worst n% of pairs based on some metric (usually point-to-point distance). \
    4.3 Rejection of pairs that are not consistent with neighboring pairs \
    4.4 Rejection of pairs containing points on mesh boundaries \
    4.5 Robust kernel outlier rejection
    
Based on tolerances wrt noise and outliers, computation cost, basin of convergence each methods have some tradeoff.

Like, using some feature edge corner we can decrease search correspondance. Similarly, normal-space sampling is better for mostly smooth area with sparse features.

Point to plane data association is used to minimize the sum of the squared distance between a point and the tangent plane at its correspondence point. But, each iteration generally slower than the point-to-point version, however, often significantly better convergence rates.

```python
import open3d as o3d
import copy
import numpy as np
import time

original_bunny = o3d.geometry.PointCloud()
moved_bunny = o3d.geometry.PointCloud()
# Goal is to move moved_bunny to match with original_bunny

bunny =  o3d.io.read_point_cloud("bunny.ply")
mesh = o3d.geometry.TriangleMesh.create_coordinate_frame(size = 0.1)
mesh_moved = o3d.geometry.TriangleMesh.create_coordinate_frame(size = 0.1)
original_bunny = bunny.voxel_down_sample(voxel_size = 0.004)

moved_bunny = copy.deepcopy(original_bunny)
R = moved_bunny.get_rotation_matrix_from_xyz((np.pi/2, np.pi/4, 0))
moved_bunny = moved_bunny.rotate(R, center=(0, 0, 0))
moved_bunny = moved_bunny.translate((0.1, 0.1, 0.1))

mesh_moved = mesh_moved.rotate( R, center = (0, 0, 0))
mesh_moved = mesh_moved.translate((0.1, 0.1, 0.1))

print(R)

vis = o3d.visualization.Visualizer()
vis.create_window()   
# o3d.visualization.draw_geometries([original_bunny, moved_bunny ,mesh ]) 
vis.add_geometry(original_bunny)
vis.add_geometry(moved_bunny)
vis.add_geometry(mesh)
vis.add_geometry(mesh_moved)

vis.update_geometry(original_bunny)
vis.update_geometry(moved_bunny)
vis.update_geometry(mesh)
vis.update_geometry(mesh_moved)

vis.poll_events()
vis.update_renderer()
time.sleep(2)
vis.destroy_window() 

def determine_correspondance(original_pcd, moved_pcd):
    correspondance_index = []
    original_pcd = np.asarray(original_pcd)
    moved_pcd = np.asarray(moved_pcd)
    original_size = np.shape(original_pcd)[0]
    moved_size = np.shape(moved_pcd)[0]
    pt2  = np.zeros( (3,1))
    for i in range(moved_size):
        pt2_temp = moved_pcd[i][:]
        pt2[0]  = pt2_temp[0] 
        pt2[1]  = pt2_temp[1] 
        pt2[2]  = pt2_temp[2] 
        dist_vec = original_pcd.T - pt2
        dist_vec = np.linalg.norm(dist_vec , axis = 0)**2
        index = np.argmin(dist_vec)
#         print( str(np.shape(dist_vec)) + " " + str(index)   )
        index_corresp = (i, index) # ( index_in_moved , index_in_original )
        correspondance_index.append(index_corresp)
    return correspondance_index

def determine_R_matrix(original_pcd, moved_pcd, correspondance_list):
    num_pairs = len(correspondance_list)
    original_pcd = np.asarray(original_pcd)
    moved_pcd = np.asarray(moved_pcd) 
    pt2_temp_original = np.zeros((3, 1))
    pt1_temp_moved = np.zeros((3, 1))
    cov = np.zeros((3, 3))
    weight = 1
    for index_in_moved, index_in_original in correspondance_list:
        pt1_temp_moved[0][0] = moved_pcd[index_in_moved][0]
        pt1_temp_moved[1][0] = moved_pcd[index_in_moved][1]
        pt1_temp_moved[2][0] = moved_pcd[index_in_moved][2]  
        pt2_temp_original[0][0] = original_pcd[index_in_original][0]
        pt2_temp_original[1][0] = original_pcd[index_in_original][1]
        pt2_temp_original[2][0] = original_pcd[index_in_original][2]
        dist = np.linalg.norm(pt1_temp_moved - pt2_temp_original)  
        if( dist >  0.02):
            weight = 1     
        cov_temp = pt2_temp_original.dot(pt1_temp_moved.T)
        cov += cov_temp*weight   
    u ,s,v = np.linalg.svd( cov)
    R = u.dot(v) 
    return R

def align_center( pcd_pts):
    center = np.mean( pcd_pts , axis =0 )
    sum_ = np.sum(pcd_pts , axis =0 )
    pcd_align = pcd_pts -center
    return pcd_align , center

def detect_stop( original_bunny_pts , pts_update ):
    value = np.linalg.norm(original_bunny_pts - pts_update)
    threshold = 3
    if( value < threshold):
        return True
    else: 
        return False

iterations = 70
vis = o3d.visualization.Visualizer()
vis.create_window() 

R_record =  np.eye(3)
for i in  range(iterations):
    original_bunny_pts , original_bunny_center = align_center( np.asarray(original_bunny.points)  )
    moved_bunny_pts , moved_bunny_center = align_center( np.asarray(moved_bunny.points)  )
    
    correspondance_list = determine_correspondance( original_bunny_pts , moved_bunny_pts )
    R= determine_R_matrix( original_bunny_pts ,moved_bunny_pts ,correspondance_list  )
    t = original_bunny_center 
    
    one = np.ones( ( np.shape(moved_bunny_pts)[0] ,1 ))
    pts = np.append(moved_bunny_pts , one , axis=1 ) 
    H = np.eye(4)
    H[0:3 , 0:3] = R
    H[0:3 , 3] = t
    H[3 , :] = [ 0,0 ,0 ,1]
    
    R_record = R @ R_record
        
    pts_update = (H @ (pts.T)).T
    pts_update = pts_update[:, 0:3]
    moved_bunny.points = o3d.utility.Vector3dVector(pts_update)
    
#     stop = detect_stop( original_bunny_pts , pts_update )
    
    moved_bunny = moved_bunny.paint_uniform_color([1, 1, 0])
    vis.add_geometry(original_bunny)
    vis.add_geometry(moved_bunny)
    vis.add_geometry(mesh)
    
    vis.update_geometry(original_bunny)
    vis.update_geometry(moved_bunny)
    vis.update_geometry(mesh)
    vis.poll_events()
    vis.update_renderer()
    time.sleep(0.01)

    
vis.destroy_window() 
o3d.visualization.draw_geometries([original_bunny, moved_bunny ,mesh ]) 

## the inverse of the rotation matrix is displayed, this demonstrates that the trasnformation
#between the original bunny and the moved bunny is determined accurately. 
print(R_record.T) 
```
